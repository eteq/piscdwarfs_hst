## Load and check out HST passbands

```python
f606_wl, f606_throughput = np.loadtxt('wfc_F606W.dat').T
f606_wl = f606_wl * u.angstrom
dwl606 = np.diff(f606_wl).mean()  # should be 1 angstrom

f814_wl, f814_throughput = np.loadtxt('wfc_F814W.dat').T
f814_wl = f814_wl * u.angstrom
dwl814 = np.diff(f814_wl).mean()  # should be 1 angstrom

plt.plot(f606_wl.value, f606_throughput, label='F606W')
plt.plot(f814_wl.value, f814_throughput, label='F814W')
plt.legend(loc=0)
plt.xlabel(r'$\lambda$')
```

```python
#one example case w/plots
T = 5800*u.K

int606 = make_color.normed_planck(f606_wl, T, fphot=True)*f606_throughput
int606e = make_color.normed_planck(f606_wl, T, fphot=False)*f606_throughput
int814 = make_color.normed_planck(f814_wl, T, fphot=True)*f814_throughput
int814e = make_color.normed_planck(f814_wl, T, fphot=False)*f814_throughput

plt.plot(f606_wl, int606, label='F606W (phot)', c='g')
plt.plot(f606_wl, int606e, label='F606W (energy)', ls=':', c='g')
plt.plot(f814_wl, int814, label='F814W (phot)', c='r')
plt.plot(f814_wl, int814e, label='F814W (energy)', ls=':', c='r')
plt.plot(f606_wl, make_color.normed_planck(f606_wl, T, fphot=True), label='BB (phot)', c='b')
plt.plot(f606_wl, make_color.normed_planck(f606_wl, T, fphot=False), label='BB (energy)', ls=':', c='b')

plt.legend(loc=0)

f1 = np.sum(int606*dwl606)
f2 = np.sum(int814*dwl814)
f1e = np.sum(int606e*dwl606)
f2e = np.sum(int814e*dwl814)
print('Fraction of photons in each band:', f1, f2, 'Energy-based', f1e, f2e)
```

```python
Ts = np.linspace(400, 100000, 1000)*u.K

int606 = make_color.normed_planck(f606_wl, Ts.reshape(-1, 1))*f606_throughput
int814 = make_color.normed_planck(f814_wl, Ts.reshape(-1, 1))*f814_throughput

plt.plot(Ts.value, np.sum(int606*dwl606,axis=1).value,label='F606W')
plt.plot(Ts.value, np.sum(int814*dwl814,axis=1).value,label='F814W')
plt.axvline(5800,color='y')
plt.legend(loc=0)

plt.xlabel('T [K]')
plt.xlim(Ts[-1].value, Ts[0].value)
plt.ylabel('frac of flux on detector')
plt.tight_layout()
```

### Make T/ratio grid

```python
T_grid, ratio_grid, leff606, leff814 = make_color.compute_T_ratio_grid(Ts, (f606_wl, f606_throughput), (f814_wl, f814_throughput))

plt.plot(T_grid.value, ratio_grid.value)
plt.axhline(1, color='k')
plt.axhline((leff814/leff606).value**4, color='r')
plt.axvline(5800,color='y')

plt.xlabel('T [K]')
plt.xlim(Ts[-1].value, Ts[0].value)
plt.ylabel('count flux ratio (f606/f814)')
plt.tight_layout()
```
