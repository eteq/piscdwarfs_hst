```python
def plot_3d(plotf, box=(0, 125)):
    fig = plt.figure(figsize=(15,15))
    ax = fig.add_subplot(111, projection='3d')
    plt.subplots_adjust(left=0,bottom=0,top=1,right=1, wspace=0, hspace=0)

    plotf(fig, ax)
    
    if box is not None:
        ax.set_xlim(*box)
        ax.set_ylim(*box)
        ax.set_zlim(*box)
    plt.tight_layout()

    @interact(azim=(-180, 180., 1.0), elev=(-90., 90., 1.0))
    def f(azim=0, elev=0):
        ax.azim = azim
        ax.elev = elev
        display.display(fig)
        
    return f
```

# IDing MW-ish analogs

```python
rng = lambda v, edge, boxsize=125: (edge<v)&(v<(boxsize-edge)) 
awayfromedges = rng(halos['x'], 15)&rng(halos['y'], 15)&rng(halos['z'], 15)
```

```python
clusterishmsk = halos['mpeak'] > 1e14
clusterishkdt = cKDTree(kdt.data[clusterishmsk])

clusterwi15msk = np.array([len(matches)>0 for matches in kdt.query_ball_tree(clusterishkdt, 15)])
clusterishkdt.data.shape, np.sum(clusterwi15msk)/len(clusterwi15msk)
```

```python
brightmsk = 1e12 < halos['mpeak']
brightkdt = cKDTree(kdt.data[brightmsk])
np.sum(brightmsk)
```

```python
mwcandidatesmsk = brightmsk & (halos['mpeak'] < 3e12) & (halos['upid']==-1) & ~clusterwi15msk & awayfromedges
np.sum(mwcandidatesmsk)
```

```python
brightids = halos['id'][brightmsk]

nnearby = []
nm31s = []

for mwcand in halos[mwcandidatesmsk]:
    xyz = [mwcand['x'], mwcand['y'], mwcand['z']]
    
    idxs = brightkdt.query_ball_point(xyz, 1)
    nmatch = len(idxs)
    if mwcand['id'] in brightids[np.array(idxs)]:
        nmatch -= 1
    nm31s.append(nmatch)
    
    idxs = brightkdt.query_ball_point(xyz, 0.4)
    nmatch = len(idxs)
    if mwcand['id'] in brightids[np.array(idxs)]:
        nmatch -= 1
    nnearby.append(nmatch)
nm31s = np.array(nm31s)
nnearby = np.array(nnearby)
mwlike = (nm31s<2)&(nnearby<1)
mwlikehalos = halos[mwcandidatesmsk][mwlike]

np.sum(nm31s<2), np.sum(nnearby<1), np.sum(mwlike), len(nm31s)
```

```python
#Double-check that this looks like a reasonable selection of mw-like thingies
def plotf(fig, ax):
    toplothalos = halos[brightmsk]
    mwlikemsk = np.in1d(toplothalos['id'], halos['id'][mwcandidatesmsk][mwlike])
    nmwlikemsk = np.in1d(toplothalos['id'], halos['id'][mwcandidatesmsk][~mwlike])

    msk1 = mwlikemsk
    msk2 = nmwlikemsk&~mwlikemsk
    msk3 = ~nmwlikemsk&~mwlikemsk
    
    ax.scatter3D(toplothalos['x'][msk1], toplothalos['y'][msk1], toplothalos['z'][msk1], color='b')
    ax.scatter3D(toplothalos['x'][msk2], toplothalos['y'][msk2], toplothalos['z'][msk2], color='r')
    ax.scatter3D(toplothalos['x'][msk3], toplothalos['y'][msk3], toplothalos['z'][msk3], color='k')

plot_3d(plotf)
```

```python
def dvs_from_host(host):
    dx = host['x'] - halos['x']
    dy = host['y'] - halos['y']
    dz = host['z'] - halos['z']
    d = np.sqrt(dx*dx + dy*dy + dz*dz)*u.Mpc
    
    dvx = host['vx'] - halos['vx']
    dvy = host['vy'] - halos['vy']
    dvz = host['vz'] - halos['vz']
    vpec = (dvx * dx + dvx * dx + dvx * dx)/d.value *u.km/u.s
    vlos = vpec + cosmology.WMAP9.H0*d
    
    s21 = bradS21*d**-2
    
    return d, vlos, s21
```

```python
galfadr1frac = 7500/41000

npiscs_s21 = defaultdict(list)
s21stocheck = [1,2,3,5,10]*u.Jy * u.km/u.s

for i, halo in enumerate(mwlikehalos[:250]):
    print(str(i) + ' of ' + str(len(mwlikehalos)), end='\r')
    d, vlos, s21 = dvs_from_host(halo)
    msk = (d > 2*u.Mpc)&(d < 10*u.Mpc)&(300*u.km/u.s<vlos)&(vlos<700*u.km/u.s)&(gkmstar<1e7*u.solMass)
    for is21 in s21stocheck:
       npiscs_s21[is21].append(np.sum(msk&(s21>is21))*galfadr1frac)


for k in sorted(npiscs_s21.keys()):
    plt.hist(npiscs_s21[k], histtype='step', label=str(k))
plt.legend(loc=0)
plt.tight_layout()
```
