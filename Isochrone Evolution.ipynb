```python
from __future__ import division, print_function

import numpy as np

from astropy import units as u
from astropy import constants as cnst
from astropy.coordinates import SkyCoord
from astropy.wcs import WCS
from astropy.io import fits, ascii
from astropy.table import Table

from astroquery.irsa_dust import IrsaDust
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
from matplotlib import rcParams

rcParams['figure.figsize'] = (16, 10)
rcParams['image.interpolation'] = 'none'
rcParams['image.origin'] = 'lower'
```

```python
def load_isoc(fn, **kwargs):
    lines = []
    headerdone = False
    with open(fn) as f:
        for l in f:
            if l.startswith('#'):
                if not headerdone and l.startswith('#      Z'):
                    lines.append('\t'.join(l[1:].strip().split()))
                    headerdone = True
            else:
                lines.append(l)
    return ascii.read(lines, **kwargs)
    
tacs = load_isoc('cmdisocs_acs_m1p0.dat')
tsdss = load_isoc('cmdisocs_sdss_m1p0.dat')
```

```python
def weighted_mag(isoc, mag):
    flux = 10**(isoc[mag]/-2.5)
    dimf = np.diff(isoc['int_IMF'])
    midflux = (flux[1:] + flux[:-1])/2
    summed_flux = np.sum(midflux*dimf)
    return -2.5*np.log10(summed_flux)

def weighted_mag_all_ages(isoc, mag):
    lages = np.unique(tsdss['log(age/yr)'])
    mags = []
    for a in lages:
        mags.append(weighted_mag(isoc[isoc['log(age/yr)']==a], mag))
    return lages, np.array(mags)
```

```python
plt.plot(*weighted_mag_all_ages(tsdss, 'mbol'), marker='o', label='mbol')
for m in 'ugriz':
    plt.plot(*weighted_mag_all_ages(tsdss, m), marker='o', label=m)
plt.legend(loc=0)
```

```python
for c in ['g-r', 'g-i', 'r-i', 'u-g', 'u-r', 'u-z']:
    m1, m2 = c.split('-')
    lages, mags1 = weighted_mag_all_ages(tsdss, m1)
    lages, mags2 = weighted_mag_all_ages(tsdss, m2)
    plt.plot(lages, mags1-mags2, marker='o', label=c)
plt.legend(loc=0)
```

```python
plt.plot(*weighted_mag_all_ages(tacs, 'mbol'), marker='o', label='mbol')
for m in ['F606W','F814W']:
    plt.plot(*weighted_mag_all_ages(tacs, m), marker='o', label=m)
plt.legend(loc=0)
```

```python
for c in ['F606W-F814W']:
    m1, m2 = c.split('-')
    lages, mags1 = weighted_mag_all_ages(tacs, m1)
    lages, mags2 = weighted_mag_all_ages(tacs, m2)
    plt.plot(lages, mags1-mags2, marker='o', label=c)
plt.legend(loc=0)
```
