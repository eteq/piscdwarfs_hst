## KK230

```python
kk230raw = np.loadtxt('hlsp_angst_hst_acs-wfc_9771-kk230_f606w-f814w_v1.phot')
kk230t = Table(data=kk230raw[:,cols.keys()],names=cols.values())
```

```python
dx = kk230t['x'] - np.median(kk230t['x'])
dy = kk230t['y'] - np.median(kk230t['y'])
dr = np.hypot(dx, dy)

plt.scatter(kk230t['F606W']-kk230t['F814W'],kk230t['F814W'], s=5, alpha=.5, edgecolor='none', c=dr*.05)
plt.colorbar().set_label('dasec')
plt.xlim(-1.5, 3.5)
plt.ylim(28,20)
plt.xlabel('F606W-F814W')
plt.ylabel('F814W')
plt.tight_layout()
```

```python
radius = 25*u.arcsec
c = kk230t['F606W']-kk230t['F814W']
m = kk230t['F814W']
clims = (.4, 1.2)
mlim = 26

plt.figure()
radmsk = dr*.05*u.arcsec < radius
plt.scatter(c[radmsk], m[radmsk], s=5, alpha=.5, edgecolor='none')
plt.xlim(-1.5, 3.5)
plt.ylim(28, 20)
plt.xlabel('F606W-F814W')
plt.ylabel('F814W')
plt.title('w/i {0}'.format(radius))
for clim in clims:
    plt.axvline(clim, c='r', ls='--')
plt.axhline(mlim, c='k', ls='--')
plt.tight_layout()
```

```python
rgbmsk = radmsk&(clims[0]<c)&(c<clims[1])&(m<mlim)&(c<4)&(m>18)
plt.hist(m[rgbmsk],bins=50, histtype='step',log=True)
xline = np.linspace(21,26,200)
plt.plot([21,26], [.2, 150], c='r')
```

# Pisc A

```python
d = np.loadtxt('PiscA_dolphot/piscA')
wcs = WCS(fits.getheader('PiscA_dolphot/PiscA_F606W_build_drc.chip1.fits'))

#object is roughly here
x0 = 3650
y0 = 5400

#control area
xc = 3500
yc = 3000

sc0 = SkyCoord.from_pixel(x0, y0, wcs)
ebvmap = IrsaDust.get_images(sc0,2*u.degree, 'ebv')[0][0].data
ebv = ebvmap[ebvmap.shape[0]//2, ebvmap.shape[0]//2]
A606W = ebv*2.471 # from S & F 11
A814W = ebv*1.526 # from S & F 11
A606W, A814W

# only use good stars
ds = d[d[:, 10]<2]

m606 = ds[:,15] - A606W
m606e = ds[:,17]
m814 = ds[:,28] - A814W
m814e = ds[:,30]
dpx_dwarf = np.hypot(x0-ds[:,2],y0-ds[:,3])
dpx_control = np.hypot(xc-ds[:,2],yc-ds[:,3])

ce = np.hypot(m606e, m814e)
me = m814e
c = m606-m814
m = m814
```

```python
nearmsk = dpx_dwarf < 500

clims = (.4, 1.5)
mlim  = 26

plt.scatter(c[nearmsk], m[nearmsk], c=dpx_dwarf[nearmsk], s=10, edgecolor='none', alpha=.7)
plt.colorbar()
plt.xlim(-1.5, 3.5)
plt.ylim(28, 20)
plt.xlabel('F606W-F814W')
plt.ylabel('F814W')

for clim in clims:
    plt.axvline(clim, color='r')

plt.axhline(mlim, color='k')
```

```python
rgbmsk = nearmsk&(clims[0]<c)&(c<clims[1])&(m<mlim)&(m>20)
plt.hist(m[rgbmsk],bins=50, histtype='step',log=True)
plt.tight_layout()
```

## No Uncs

```python
mod = rgbmcmc.RGBModel(m[rgbmsk], tipprior=(23, 25.5))

st = time.time()
sampler, res = mod.make_and_run_sampler(niters=1500, burnin=500,tipmag0=25)
et = time.time()
print('Took',et-st,'sec')

triangle.corner(sampler.flatchain, labels=mod.PARAM_NAMES)
None
```

```python
tm = sampler.flatchain[:,0]
plt.hist(tm,bins=100, histtype='step')
yls = plt.ylim()
plt.plot(np.sort(tm), np.linspace(yls[0],yls[1],len(tm)))
p50y = (yls[0]+yls[1])/2
plt.axhline(p50y, c='r', ls='--')
med, m90, p90 = np.percentile(tm, [50,10,90])
plt.axvline(med, c='r', ls=':')
plt.tight_layout()
print(med, m90, p90, med-m90, p90-med)

plt.figure()
plt.scatter(c[rgbmsk], m[rgbmsk], s=5, alpha=.5, edgecolor='none')
plt.axhline(med,c='r')
plt.axhline(m90,c='r',ls='--')
plt.axhline(p90,c='r',ls='--')
plt.xlim(-1.5, 3.5)
plt.ylim(28, 20)
plt.xlabel('F606W-F814W')
plt.ylabel('F814W')
plt.tight_layout()
```
